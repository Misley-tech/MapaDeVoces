```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# Load the audio file
path_voz_AnaPrueba10dB = 'C:/Users/Asus/OneDrive/Escritorio/MapaDeVoces/VocesAlamacenadas/VocesGrabadasSeba/AnaPrueba10dB.wav'
path_voz_AnaPrueba6dB = 'C:/Users/Asus/OneDrive/Escritorio/MapaDeVoces/VocesAlamacenadas/VocesGrabadasSeba/AnaPrueba6dB.wav'
path_voz_AnaPrueba = 'C:/Users/Asus/OneDrive/Escritorio/MapaDeVoces/VocesAlamacenadas/VocesGrabadasSeba/AnaPrueba.wav'
y, sr = librosa.load(path_voz_AnaPrueba10dB)
# y, sr = librosa.load(path_voz_AnaPrueba6dB)
# y, sr = librosa.load(path_voz_AnaPrueba)

# Extract MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=15)  # You can change n_mfcc to your desired number of coefficients

# Display MFCCs
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
